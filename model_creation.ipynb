### Seperating the features and the target 

```python
X = df.drop('target', axis='columns') 
y = df.target 
```

```python
X.head() 
```

```python
y 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

### Standarlize the data 

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()  
X_train_std = scaler.fit_transform(X_train) 
X_test_std = scaler.fit_transform(X_test) 
```

### Building the neural network 

```python
import tensorflow as tf  
from tensorflow import keras
tf.random.set_seed(3) 
```

```python
# set the layers of the neural network 
model = keras.Sequential([
    # flatten to convert the data into single dimensional array, 30 because there are 30 columns 
    keras.layers.Flatten(input_shape=(30,)), 
    # there are 20 neurons in the hidden layer
    keras.layers.Dense(20, activation='relu'), 
    # output layer, 2 neurons because target has 2 classes
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network 
model.compile(
    optimizer= 'adam', 
    # when we use one hot encode we should use categorical_crossentropy 
    # when there are few lables we can use sparse categorical crossentropy
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
) 
```

```python
# training the neural network 
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=20)  
```

### Visualizing the accuracy and loss 

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy']) 

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs') 

plt.legend(['training data', 'validationd data'], loc = 'lower right') 
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss']) 

plt.title('model loss') 
plt.ylabel('loss')
plt.xlabel('epochs') 

plt.legend(['training data', 'validation data'], loc = 'upper right') 
```

```python
# Accuracy of the model data 
loss, accuracy = model.evaluate(X_test_std, y_test) 
accuracy 
```
